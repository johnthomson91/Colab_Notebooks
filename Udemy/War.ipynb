# Card Class

```
class Card():

  def __init__(self, suit, rank):
    self.suit = suit
    self.rank = rank
    self.value = values[rank]

  def __str__(self):
    return f"{self.rank} of {self.suit}"
```

# Deck Class


```
class Deck:

  def __init__(self):
    self.all_cards = []

    for suit in suits:
      for i in rank:

        card_created = Card(suit, i)

        self.all_cards.append(card_created)


  def shuffle(self):
    random.shuffle(self.all_cards)


  def deal_card(self):
    return self.all_cards.pop()
```

# Player Class

```
class Player:
  def __init__(self, name):
    self.name = name
    self.all_cards = []
  
  def remove_cards(self):
    # take the "top" card from players deck to be played vs opponent
    return self.all_cards.pop(0)

  def add_cards(self, new_cards):

    if type(new_cards) == list:
      # because .append doesnt work for lists or multiple cards 
      # we have to use .extend() here
      self.all_cards.extend(new_cards)
    else:
      # if new_cards is just a single card we can use .append()
      self.all_cards.append(new_cards)
  
  def __str__(self):
    return f"Player {self.name} has {len(self.all_cards)} cards"
```
