# Timing Your Python Code

```
def func1(n):
  return [str(num) for num in range(n)]
```

```
def func2(n):
  return list(map(str,range(n)))
```

```
import time

start_time = time.time()
result  = func1(1000000)
end_time = time.time()
elapsed_time = end_time - start_time

print(elapsed_time)
```

```
import time

start_time = time.time()
result  = func2(1000000)
end_time = time.time()
elapsed_time = end_time - start_time

print(elapsed_time)
```

```
import timeit
```

```
stmt1 = '''
func1(100)
'''
```

```
setup1 = '''
def func1(n):
  return [str(num) for num in range(n)]
'''
```

```
timeit.timeit(stmt1, setup1, number=1000000)
```

```
stmt2 = '''
func2(100)
'''
```

```
setup2 = '''
def func2(n):
  return list(map(str,range(n)))
'''
```

```
timeit.timeit(stmt2, setup2, number=1000000)
```
