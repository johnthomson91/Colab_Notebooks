# vowelCount(s) [15 pts]
Write the function vowelCount(s), that takes a string s, and returns the number of vowels in s, ignoring case, so "A" and "a" are both vowels. 

The vowels are "a", "e", "i", "o", and "u". 

So, for example:
assert(vowelCount("Abc def!!! a? yzyzyz!") == 3) # two a's and one e

```
def vowelCount(some_string):

  some_string = some_string.lower()
  vowels = 'aeiou'
  count = 0

  for letter in some_string:
    if letter in vowels:
      count += 1
  
  return count
  
```

```
vowelCount("Abc def!!! a? yzyzyz!")
```

```
vowelCount("Hello World, my government name is John Anthony Thomson")
```

# applyCaesarCipher(message, shift) [20 pts]
A Caesar Cipher is a simple cipher that works by shifting each letter in the given message by a certain number. For example, if we shift the message "We Attack At Dawn" by 1 letter, it becomes "Xf Buubdl Bu Ebxo".

Write the function applyCaesarCipher(message, shift) which shifts the given message by shift letters. You are guaranteed that message is a string, and that shift is an integer between -25 and 25. Capital letters should stay capital and lowercase letters should stay lowercase, and non-letter characters should not be changed. Note that "Z" wraps around to "A". So, for example:

```
def applyCaesarCipher(message, shift):

  alphabet_lower = "abcdefghijklmnopqrstuvwxyz"
  alphabet_upper = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

  translated_message = ""

  for letter in message:
    if letter in alphabet_lower:
        letter_value = alphabet_lower.find(letter)
        translated_message += alphabet_lower[(letter_value + shift) % 26]      
    elif letter in alphabet_upper:
        letter_value = alphabet_upper.find(letter)
        translated_message += alphabet_upper[(letter_value + shift) % 26]
    else:
        translated_message += letter
  
  return translated_message
```

```
 applyCaesarCipher("We Attack At Dawn", 1)
```

```
applyCaesarCipher("zodiac", -2)
```

# rotateStringLeft(s, n) [20 pts]

Write the function rotateStringLeft(s, n) that takes a string s and a possibly-negative integer n. If n is non-negative, the function returns the string s rotated n places to the left. If n is negative, the function returns the string s rotated |n| places to the right. 

So, for example:

assert(rotateStringLeft('abcd',  1) == 'bcda')

assert(rotateStringLeft('abcd', -1) == 'dabc')

```
def rotateStringLeft(s,n):
    return s[n:] + s[0:n]
    
```

```
rotateStringLeft('abcd',1)
```

```
rotateStringLeft('123456789', 3)
```

# Right-Left Route Ciphers (encode + decode) [35 pts]

Background: A right-left route cipher is a fairly simple way to encrypt a message. It takes two values, some plaintext and a number of rows, and it first constructs a grid with that number of rows and the minimum number of columns required, writing the message in successive columns

```
    For example, if the message is WEATTACKATDAWN, with 4 rows, the grid would be:
    
    W T A W
    E A T N
    A C D 
    T K A 
```

```
def encodeRightLeftRouteCipher(some_string, grid_rows):

  cipher = ''

  for row in range(grid_rows):
    if row % 2 == 0:
      cipher += (some_string[row::grid_rows])
    else:
      temp = (some_string[row::grid_rows])
      cipher += temp[::-1]


  print(f'{grid_rows}{cipher}')
  
```
