# for loops and range

```
def sumFromMtoN(m,n):
  return sum(range(m,n+1))
```

```
sumFromMtoN(5,10) # 5+6+7+8+9+10
```

```
def sumEveryKthFromMtoN(m, n, k):
  total = 0

  for x in range(m, n+1, k):
    total += x
   
  return total
```

```
sumEveryKthFromMtoN(5 , 20, 7) == 5 + 12 + 19
```

```
def sumToN(n):
  return sum(range(n+1))
```

```
sumToN(5) # 0+1+2+3+4+5
```

```
def sumOfOddsFromMtoN(m , n):
  total = 0

  for x in range(m, n+1):
    if x%2 == 1:
      total += x
  return total
```

```
sumOfOddsFromMtoN(4, 10) # 5 + 7 + 9
```

# nested loops

```
# we can put loops inside of loops to repeat actions at multiple levels
# this prints coordinates

def printCoordinates(xMax, yMax):
  for x in range(xMax + 1):
    for y in range(yMax + 1):
      print(f"({x}, {y}), " , end="")
    print()
```

```
printCoordinates(4,5)
```

```
def printCoordinates2(xMax, yMax):
  for x in range(xMax + 1):
    for y in range(x + 1):
      print(f"({x}, {y}) , " , end="")
    print()
```

```
printCoordinates2(4,5)
```

find the nth number that is a multiple of either 4  or 7

```
def isMultiple4or7(x):
  return x%4 == 0  or x%7 == 0

def nthMultiple4or7(n):
  found = 0
  guess = 0

  while found < n:
    guess +=1
    if isMultiple4or7(guess):
      found +=1

  return guess

print("Multiples of 4 or 7: ", end="")
for n in range(15):
    print(nthMultiple4or7(n), end=" ")
print()
```

# break and continue


```
for n in range(200):
  if n%3 == 0:
    continue # skips the rest of this pass (skips everything)
  elif n == 8:
    break # we're done- stops the entire loop once we hit 8 even though we're going up to 200
  print(n, end=' ')
```

```
for n in range(200):
  if n%3 == 0:
    continue # skips the rest of this if statement, goes to next numer in the range
  elif n == 31:
    break # we're done- stops the entire loop once we hit 31 even though we're going up to 200
  else:
    pass # does nothing- pass is a place holder here
  print(n, end=' ')
```

# isPrime


```
def isPrime(n):
  if n < 2:
    return False
  for factor in range(2, n):
    if n % factor == 0:
      return False
  else:
    return True

for i in range(100):
  if isPrime(i):
    print(i , end= ' ')
print()
```

```
import math

def isPrime(n):
  if n < 2:
    return False
  elif n == 2:
    return True
  elif n % 2 == 0:
    return False

  maxFactor = int(math.sqrt(n))
  for factor in range(3, maxFactor + 1, 2):
    if n % factor == 0:
      return False
  else:
    return True

# given n, nth prime number will come after n (i.e. n+1)


def nthPrime(n):
  found = 0
  guess = 0

  print(n, found, guess)
  while found <= n:
    guess += 1
    if isPrime(guess):
      
      found += 1
  return guess


for i in range(10):
  print(i, nthPrime(i))
print('Done!')
```
